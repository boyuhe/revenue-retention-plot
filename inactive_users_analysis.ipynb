```python
from lifelines import  KaplanMeierFitter
import pandas as pd
import numpy as np
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
import os
os.chdir('/Users/magictavern/Downloads')
```

```python
df=pd.read_csv('1546081431900.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
ax = subplot(111)
t = np.linspace(0, 50, 51)
kmf = KaplanMeierFitter()
kmf.fit(df['duration'], event_observed=df['return'], timeline=t, label="all")
ax = kmf.plot(ax=ax,figsize=(20,10))
plt.title("not return rate")
```

```python
ax = subplot(111)
t = np.linspace(0, 50, 51)
ax = (1 - kmf.survival_function_).plot(ax=ax,figsize=(20,10))
plt.title("return rate")
```
